```python
import numpy as np
import random
import matplotlib.pyplot as pt
import math


```

```python

m=1000 # number of points

xr=list(10*np.random.rand(m)) # x red-coordinates of points
yr=list(20*np.random.rand(m)) # y- red _coordinates of points
xb=list(20*np.random.rand(m))   # x blue-coordinates of points
yb=list(10*np.random.rand(m))   # y blue-coordinates of points
pt.scatter(xr,yr, color='red')
pt.scatter(xb,yb, color='blue')

```

```python
xt=5.0 #new x data point
yt=5.0 # new y data point 
pt.scatter(xt,yt,s=200,marker='*',color='y')
pt.show()
```

```python
dr=[] # distance from red
for xrr,yrr in zip(xr,yr):
                    dr.append(math.sqrt((xrr-xt)**2+(yrr-yt)**2))
db=[]# distance from blue 
for xbb,ybb in zip(xb,yb):
                    db.append(math.sqrt((xbb-xt)**2+(ybb-yt)**2))
drsort=sorted(dr)
dbsort=sorted(db)
```

```python
duri={'r0':drsort[0],'r1':drsort[1],'r2':drsort[2],'r3':drsort[3],'r4':drsort[4],'b0':dbsort[0],'b1':dbsort[1],'b2':dbsort[2],'b3':dbsort[3],'b4':dbsort[4]}
rizband=sorted(duri.items(),key=lambda item:item[1]) 
```

```python
rizband
```

```python

riz=[]  
for zz in rizband: 
    riz.append(zz[0]) 
rizz=[] 
for i in range(5): 
    rizz.append(riz[i][0][0])
rr=0
for rb in rizz:
    if rb=='r':
        rr+=1
if rr>=4:
    rang='red'
else:
    rang='blue'
```

```python
pt.scatter(xr,yr, color='red')
pt.scatter(xb,yb, color='blue')
pt.scatter(xt,yt,s=1000,marker='*',color=rang)
pt.axis('equal')
```
